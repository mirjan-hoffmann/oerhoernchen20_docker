```python
r = requests.get(gsheet_link)
```

```python
r_json = r.json()
print(json.dumps(r_json, indent=2))
```

```python
for item in r_json['feed']['entry']:
#     pprint.pprint(item)
    entry['name'] = item['gsx$titel']['$t']
    entry['url'] = item['gsx$url']['$t']
    entry['about'] = item['gsx$beschreibung']['$t']
    pprint.pprint(entry)
```
